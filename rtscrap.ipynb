```python
filename="RTNews.csv"
f=open(filename,"w", encoding = 'utf-8')
headers="Statement, Link, Date \n"
f.write(headers)
frame=[]

data = BeautifulSoup(driver.page_source, 'html.parser')
links=data.find_all('div',attrs={'class':'list-card__content'})
# links
for j in links:
#     print(j)
#     print("hello")
    Statement = j.find("div",attrs={'class':'list-card__content--title'})#.find('a').text.strip()  Link+
    if Statement==None:
        continue
    Headline = Statement.find('a').text.strip()
    Link = "https://www.rt.com"
    Link += Statement.find('a')['href'].strip()
#     print(Link)
#     print(Headline)
    Date = j.find('div',attrs={'class':'card__date'})
    
    if Date==None:
        continue
    DateTime = Date.find('span').text.strip()
    print(DateTime)
    frame.append((Headline,Link,DateTime))
#     f.write(Statement.replace(",","^")+","+Link+","+Date.replace(",","^")+","+"\n")
    
    
    
    
f.close()

data=pd.DataFrame(frame, columns=['Statement','Link','Date'])
data.head()

#     DateTime = Date.find('span').text.strip()
#     print(DateTime)
```
